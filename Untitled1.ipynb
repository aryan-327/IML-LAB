```python
import numpy as np

x = np.array([[1,1],[1,2],[1,3]])
y = np.array([[2,3],[4,5],[6,7]])
x_t=x.T
x_t_x=np.dot(x_t,x)
x_t_y=np.dot(x_t,y)
inv_xt_x=np.linalg.inv(x_t_x)
w=np.dot(inv_xt_x,x_t_y)
#print(w)

print(f'y1={w[0][0]:.2f} + {w[0][1]:.2f}x')
print(f'y2={w[1][0]:.2f} + {w[1][1]:.2f}x')

```

# Task
Perform multivariate linear regression to model 'y1' and 'y2' based on 'x' using the provided values (x=[1,2,3], y1=[2,4,6], y2=[3,5,7]), generate the regression equations, calculate and display essential error metrics (e.g., Mean Squared Error (MSE), Root Mean Squared Error (RMSE), and R-squared) for each dependent variable, and provide a summary of the derived equations, error metrics, and insights into the model's performance.

## Create Dataset from Provided Values

### Subtask:
Construct a pandas DataFrame using the provided 'x', 'y1', and 'y2' values. Specifically, 'x' will be [1,2,3], 'y1' will be [2,4,6], and 'y2' will be [3,5,7].


**Reasoning**:
To construct the DataFrame as specified, I will first import pandas, then create a dictionary with the given 'x', 'y1', and 'y2' values, and finally convert this dictionary into a pandas DataFrame and display its head.



```python
import pandas as pd

data = {
    'x': [1, 2, 3],
    'y1': [2, 4, 6],
    'y2': [3, 5, 7]
}

df = pd.DataFrame(data)
print("DataFrame created:")
print(df.head())
```

## Prepare Data for Multivariate Regression

### Subtask:
Define the independent variable (X) as 'x' and the dependent variables (Y) as 'y1' and 'y2'. Then, split this data into training and testing sets to properly evaluate the model for each dependent variable.


**Reasoning**:
To prepare the data for multivariate regression, I need to define the independent variable (X) with an added constant term for the intercept, combine the dependent variables (Y) 'y1' and 'y2', and then split these into training and testing sets to evaluate the model.



```python
import numpy as np
from sklearn.model_selection import train_test_split

# 1. Extract the independent variable 'x' and add a constant term
X = df[['x']].values
X = np.c_[np.ones(X.shape[0]), X] # Add a column of ones for the intercept

# 2. Extract the dependent variables 'y1' and 'y2' and combine them into a single DataFrame Y
Y = df[['y1', 'y2']]

# 3. and 4. Split X and Y into training and testing sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.33, random_state=42)

# 5. Print the shapes of the resulting datasets
print("Shape of X_train:", X_train.shape)
print("Shape of X_test:", X_test.shape)
print("Shape of Y_train:", Y_train.shape)
print("Shape of Y_test:", Y_test.shape)
```

## Build and Train Multivariate Regression Model

### Subtask:
Initialize and train a multivariate linear regression model. This will involve fitting the model to determine the coefficients and intercept for 'x' for both 'y1' and 'y2' predictions simultaneously or as separate models.


**Reasoning**:
To initialize and train a multivariate linear regression model, I will import `LinearRegression` from `sklearn.linear_model`, instantiate it, and then fit the model using the prepared training data `X_train` and `Y_train`.



```python
from sklearn.linear_model import LinearRegression

# Initialize the Linear Regression model
model = LinearRegression()

# Fit the model to the training data
model.fit(X_train, Y_train)

print("Multivariate Linear Regression model trained successfully.")
```

**Reasoning**:
With the model trained, the next step is to make predictions on the test data (`X_test`) to evaluate the model's performance on unseen data. This will provide the predicted values for 'y1' and 'y2' which are necessary for calculating error metrics.



```python
Y_pred = model.predict(X_test)
print("Predictions made successfully.")
```

**Reasoning**:
To evaluate the model's performance, I will calculate essential error metrics such as Mean Squared Error (MSE), Root Mean Squared Error (RMSE), and R-squared for both 'y1' and 'y2' using the actual test data and the predictions.



```python
from sklearn.metrics import mean_squared_error, r2_score

# Calculate metrics for y1
mse_y1 = mean_squared_error(Y_test['y1'], Y_pred[:, 0])
rmse_y1 = np.sqrt(mse_y1)
r2_y1 = r2_score(Y_test['y1'], Y_pred[:, 0])

# Calculate metrics for y2
mse_y2 = mean_squared_error(Y_test['y2'], Y_pred[:, 1])
rmse_y2 = np.sqrt(mse_y2)
r2_y2 = r2_score(Y_test['y2'], Y_pred[:, 1])

print(f"--- Metrics for y1 ---")
print(f"Mean Squared Error (MSE): {mse_y1:.4f}")
print(f"Root Mean Squared Error (RMSE): {rmse_y1:.4f}")
print(f"R-squared (R2): {r2_y1:.4f}")
print(f"\n--- Metrics for y2 ---")
print(f"Mean Squared Error (MSE): {mse_y2:.4f}")
print(f"Root Mean Squared Error (RMSE): {rmse_y2:.4f}")
print(f"R-squared (R2): {r2_y2:.4f}")
```

**Reasoning**:
The R-squared calculation produced 'nan' and an 'UndefinedMetricWarning' because the `train_test_split` function, when applied to a dataset of only three samples with a test size of 0.33, resulted in a test set with only one sample. R-squared requires at least two samples to be well-defined. To resolve this for such a small dataset and allow all metrics to be calculated for demonstration purposes, I will modify the data preparation step to use the entire dataset for both training and testing.



```python
import numpy as np

# 1. Extract the independent variable 'x' and add a constant term
X = df[['x']].values
X = np.c_[np.ones(X.shape[0]), X] # Add a column of ones for the intercept

# 2. Extract the dependent variables 'y1' and 'y2' and combine them into a single DataFrame Y
Y = df[['y1', 'y2']]

# For this small dataset, use the entire dataset as both training and testing set
# to avoid UndefinedMetricWarning for R-squared with less than two samples.
X_train, X_test = X, X
Y_train, Y_test = Y, Y

# 5. Print the shapes of the resulting datasets
print("Shape of X_train:", X_train.shape)
print("Shape of X_test:", X_test.shape)
print("Shape of Y_train:", Y_train.shape)
print("Shape of Y_test:", Y_test.shape)
```

**Reasoning**:
Since the data preparation step was modified to use the entire dataset for both training and testing, the model needs to be re-trained with these updated `X_train` and `Y_train` datasets to ensure consistency before re-evaluating its performance.



```python
from sklearn.linear_model import LinearRegression

# Initialize the Linear Regression model
model = LinearRegression()

# Fit the model to the training data
model.fit(X_train, Y_train)

print("Multivariate Linear Regression model re-trained successfully with full dataset.")
```

**Reasoning**:
With the model re-trained on the full dataset, I need to re-run the prediction step to generate new predictions (`Y_pred`) using the updated `X_test` (which now also represents the full dataset).



```python
Y_pred = model.predict(X_test)
print("Predictions made successfully with re-trained model.")
```

**Reasoning**:
Now that the model has been re-trained and new predictions have been made on the full dataset, I will re-calculate the error metrics (MSE, RMSE, and R-squared) for both 'y1' and 'y2' using the updated `Y_test` and `Y_pred` to reflect the changes and obtain valid R-squared values.



```python
from sklearn.metrics import mean_squared_error, r2_score

# Calculate metrics for y1
mse_y1 = mean_squared_error(Y_test['y1'], Y_pred[:, 0])
rmse_y1 = np.sqrt(mse_y1)
r2_y1 = r2_score(Y_test['y1'], Y_pred[:, 0])

# Calculate metrics for y2
mse_y2 = mean_squared_error(Y_test['y2'], Y_pred[:, 1])
rmse_y2 = np.sqrt(mse_y2)
r2_y2 = r2_score(Y_test['y2'], Y_pred[:, 1])

print(f"--- Metrics for y1 ---")
print(f"Mean Squared Error (MSE): {mse_y1:.4f}")
print(f"Root Mean Squared Error (RMSE): {rmse_y1:.4f}")
print(f"R-squared (R2): {r2_y1:.4f}")
print(f"\n--- Metrics for y2 ---")
print(f"Mean Squared Error (MSE): {mse_y2:.4f}")
print(f"Root Mean Squared Error (RMSE): {rmse_y2:.4f}")
print(f"R-squared (R2): {r2_y2:.4f}")
```

**Reasoning**:
With the model re-trained and error metrics calculated, the next step is to extract and display the regression equations (coefficients and intercept for 'y1' and 'y2') and provide a summary of all derived information, including the error metrics.



```python
print("\n--- Derived Regression Equations ---")

# The intercept is shared across all dependent variables in a single LinearRegression model
intercept = model.intercept_

# The coefficients correspond to the features in X_train.
# Since X_train was np.c_[np.ones(X.shape[0]), X], the first coefficient is for the constant (which is handled by intercept),
# and the second is for 'x'.
coefficients = model.coef_[:, 1] # Only the coefficient for 'x' (the second column in coef_)

# For y1
print(f"Equation for y1: y1 = {intercept[0]:.2f} + {coefficients[0]:.2f}x")

# For y2
print(f"Equation for y2: y2 = {intercept[1]:.2f} + {coefficients[1]:.2f}x")

print("\n--- Summary of Model Performance ---")
print(f"For y1:")
print(f"  Mean Squared Error (MSE): {mse_y1:.4f}")
print(f"  Root Mean Squared Error (RMSE): {rmse_y1:.4f}")
print(f"  R-squared (R2): {r2_y1:.4f}")

print(f"\nFor y2:")
print(f"  Mean Squared Error (MSE): {mse_y2:.4f}")
print(f"  Root Mean Squared Error (RMSE): {rmse_y2:.4f}")
print(f"  R-squared (R2): {r2_y2:.4f}")

print("\n--- Insights ---")
print("The model perfectly fits the data (MSE=0, R2=1) for both y1 and y2. This is expected given the small, perfectly linear dataset and using the entire dataset for training and testing. In real-world scenarios, a perfect fit is rare and usually indicates overfitting if not carefully validated.")
```
