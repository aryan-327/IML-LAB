```python
#prime no
n = int(input('enter a number: '))
c=0
if n ==0:
  print('0 is neither positive nor negative')
elif  n<0:
  print('enter positive number')
else:
 for i in range (1,n+1):
   if n%i==0:
    c+=1
 if c ==2:
    print(n, ' is prime')
 else:
    print(n, ' is not prime')
```

```python
# check an integer is positive , negative or zero without using if else
def check(x,y):
  return ( x>y) - (x<y)          # true is 1  and false is 0

def answer(n):
  return{-1:'nagative', 0:'zero',1:'positive'}[check(n,0)]

n = int(input('enter a number: '))
print(n,' is ',answer(n) )



```
