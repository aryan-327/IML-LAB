```python

#single inheritance

class p :
    def parent(self):
        print('parent method')
class c (p):
    def child(self):
        print("child method")

child_obj = c()
child_obj.child()
child_obj.parent()

```

```python

#multi level inheritance
class grandparent :
    def property(self):
        print('grand parent method')
class parent (grandparent):
    def business(self):
        print("parent method")

class child(parent):
    def education(self):
        print('child  method')



print("\nchild accessing child , parent and grandparent")
c_obj = child()
c_obj.property()
c_obj.business()
c_obj.education()


```

```python



#multiple inheritance

class Father:
    def skill1(self):
        print('skill1')
class mother:
    def skill2(self):
        print('skill2')

class child (Father,mother):
    def skill3(self):
        print('skill3')

father_obj=Father()
mother_obj = mother()
child_obj=child()

print("\nchild accessing father")
child_obj.skill1()

print("\nchild accessing mother")
child_obj.skill2()
print("\nchild accessing child function")
child_obj.skill3()
```

```python


#overriding
class parent:
    def speak(self):
        print("parent speak")

class child(parent):
    def speak(self):
        print("child speak")

c = child()
c.speak()

p = parent()
p.speak()

```

```python

#overloading
class parent:
    def speak(self,a='hi'):
        print("parent speak" ,a)


p = parent()
p.speak()
a='study'
p.speak(a)

```

```python

#operator overloading
class n:
    def __init__(self , value):
        self.value=value
    def __add__(self,other):
        return n(self.value + other.value)

    def display(self):
        print("Self: ",self.value)

n1 = n(10)
n2 = n(20)
n3 = n2+n1
print(n3)



```
