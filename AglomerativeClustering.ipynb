```python
import numpy as np
from scipy.cluster.hierarchy import linkage, dendrogram, fcluster
import matplotlib.pyplot as plt

# Data points
points = np.array([
    [1, 1],  # A
    [2, 1],  # B
    [4, 3],  # C
    [5, 4]   # D
])

labels = ['A', 'B', 'C', 'D']

# Perform Agglomerative Clustering (Single Linkage)
Z = linkage(points, method='single')

# Plot dendrogram
plt.figure()
dendrogram(Z, labels=labels)
plt.title("Dendrogram")
plt.xlabel("Points")
plt.ylabel("Distance")
plt.show()

# Form 2 clusters
clusters = fcluster(Z, 2, criterion='maxclust')

print("Cluster Assignments:")
for label, cluster in zip(labels, clusters):
    print(label, "-> Cluster", cluster)
```

##Interpretation
Points A(1,1) and B(2,1) are very close → merged first.
Points C(4,3) and D(5,4) are also close → merged second.
Two natural clusters formed:
Cluster 1: {A, B}
Cluster 2: {C, D}

This shows data is clearly divided into two spatial groups.

##Conclusion
Agglomerative clustering successfully grouped the data based on similarity.
Final clusters:
{A, B}
{C, D}
