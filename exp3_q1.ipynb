```python
import numpy as np

X = np.array([[1,2],[1,3],[1,4]])
y = np.array([[1],[2],[3]])

print("Design matrix X:")
print(X)
print("\nTarget vector y:")
print(y)
```

```python
X_transpose = X.T
Xt_X = np.dot(X_transpose, X)

lambda_val = 1

# User-defined 'I' matrix
I = np.array([[0,0],[0,1]])

regularization_term = lambda_val * I

regularized_term = Xt_X + regularization_term

print("X_transpose:")
print(X_transpose)
print("\nXt_X:")
print(Xt_X)
print("\nLambda value:", lambda_val)
print("\nUser-defined matrix I:")
print(I)
print("\nRegularization term (lambda * I):")
print(regularization_term)
print("\nRegularized term (Xt_X + lambda * I):")
print(regularized_term)
```

```python
inverse_regularized_term = np.linalg.inv(regularized_term)

print("Inverse of (Xt_X + lambda * I):")
print(inverse_regularized_term)
```

```python
Xt_y = np.dot(X_transpose, y)

print("X_transpose * y (Xt_y):")
print(Xt_y)
```

```python
ridge_coefficients = np.dot(inverse_regularized_term, Xt_y)

print("Ridge Regression Coefficients (beta):")
# Display coefficients in decimal format without scientific notation
np.set_printoptions(suppress=True, formatter={'float_kind':'{:0.8f}'.format})
print(ridge_coefficients)
```
