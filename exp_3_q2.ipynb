# Task
Calculate Ridge Regression coefficients for the given design matrix X = `[[1,2,3],[2,3,2],[3,4,3]]` and target vector y = `[[1],[2],[3]]` using a regularization parameter lambda = 1. Display the calculated coefficients in decimal format without scientific notation and summarize their implications.

## Define X and y

### Subtask:
Define the design matrix X as `[[1,2,3],[2,3,2],[3,4,3]]` and the target vector y as `[[1],[2],[3]]`.


**Reasoning**:
The subtask requires defining the design matrix X and target vector y using NumPy. I will import NumPy and then create the arrays X and y as specified.



```python
import numpy as np

X = np.array([[1,1,2],[1,2,3],[1,3,4]])
y = np.array([[1],[2],[3]])

print("Design matrix X:")
print(X)
print("\nTarget vector y:")
print(y)
```

## Calculate (X^T * X + lambda * I)

### Subtask:
Calculate the product of the transpose of X and X, then add the regularization term (lambda * I), where lambda = 1 and I is an identity matrix of the appropriate size.


**Reasoning**:
I need to calculate (X^T * X + lambda * I) by first calculating the transpose of X, then the dot product of X transpose and X, defining lambda, creating an identity matrix, multiplying lambda by the identity matrix, and finally adding this to the dot product, as per the instructions.



```python
X_transpose = X.T
Xt_X = np.dot(X_transpose, X)

lambda_val = 1

I = np.identity(X.shape[1])

regularization_term = lambda_val * I

regularized_term = Xt_X + regularization_term

print("X_transpose:")
print(X_transpose)
print("\nXt_X:")
print(Xt_X)
print("\nLambda value:", lambda_val)
print("\nIdentity matrix I:")
print(I)
print("\nRegularization term (lambda * I):")
print(regularization_term)
print("\nRegularized term (Xt_X + lambda * I):")
print(regularized_term)
```

**Reasoning**:
The next step in calculating Ridge Regression coefficients is to find the inverse of the `regularized_term` matrix. I will use `np.linalg.inv()` for this purpose.



```python
inverse_regularized_term = np.linalg.inv(regularized_term)

print("Inverse of (Xt_X + lambda * I):")
print(inverse_regularized_term)
```

# Task
The next step is to calculate `X^T * y`. This is done by taking the dot product of the transpose of the design matrix `X` and the target vector `y`. This calculation is a necessary component in the formula for Ridge Regression coefficients.

## Calculate X^T * y

### Subtask:
Calculate the product of the transpose of X and the target vector y.


**Reasoning**:
The subtask requires calculating the dot product of the transpose of X (`X_transpose`) and the target vector y, and then printing the result. I will use `np.dot()` for the calculation and store it in `Xt_y`.



```python
Xt_y = np.dot(X_transpose, y)

print("X_transpose * y (Xt_y):")
print(Xt_y)
```

**Reasoning**:
I need to calculate the Ridge Regression coefficients (beta) by multiplying the inverse of the regularized term `inverse_regularized_term` with the product of the transpose of X and y (`Xt_y`), then print the result and format it to decimal without scientific notation as requested in the main task.



```python
ridge_coefficients = np.dot(inverse_regularized_term, Xt_y)

print("Ridge Regression Coefficients (beta):")
# Display coefficients in decimal format without scientific notation
np.set_printoptions(suppress=True, formatter={'float_kind':'{:0.8f}'.format})
print(ridge_coefficients)
```
