```python
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Define the data points
data_points = np.array([
    [2, 3],  # A
    [3, 4],  # B
    [6, 6],  # C
    [7, 7]   # D
])

# Initial centroids
# According to the problem, initial centroids are A(2,3) and C(6,6)
initial_centroids = np.array([
    [2, 3],
    [6, 6]
])

print("Data Points:")
for i, point in enumerate(data_points):
    print(f"Point {chr(65+i)}: {point}")
print("\nInitial Centroids:")
print(f"Centroid 1 (A): {initial_centroids[0]}")
print(f"Centroid 2 (C): {initial_centroids[1]}")
```

```python
# Initialize KMeans with the specified number of clusters (k=2) and initial centroids
# We set n_init=1 to ensure that the algorithm uses the provided initial centroids exactly once
kmeans = KMeans(n_clusters=2, init=initial_centroids, n_init=1, random_state=42)

# Fit the model to the data points
kmeans.fit(data_points)

# Get the cluster assignments for each data point
labels = kmeans.labels_

# Get the final centroids
final_centroids = kmeans.cluster_centers_

print("\nCluster Assignments:")
for i, point_label in enumerate(labels):
    print(f"Point {chr(65+i)} ({data_points[i]}): Cluster {point_label}")

print("\nFinal Centroids:")
for i, centroid in enumerate(final_centroids):
    print(f"Centroid {i}: {centroid}")

# Divide the dataset into 2 parts
cluster_0 = data_points[labels == 0]
cluster_1 = data_points[labels == 1]

print("\nDataset Divided into 2 Parts:")
print(f"Cluster 0: {cluster_0}")
print(f"Cluster 1: {cluster_1}")
```

```python
# Visualize the clusters
plt.figure(figsize=(8, 6))

plt.scatter(data_points[labels == 0, 0], data_points[labels == 0, 1], s=100, c='red', label='Cluster 0')
plt.scatter(data_points[labels == 1, 0], data_points[labels == 1, 1], s=100, c='blue', label='Cluster 1')

plt.scatter(final_centroids[:, 0], final_centroids[:, 1], s=200, c='black', marker='X', label='Final Centroids')

# Annotate points
for i, point in enumerate(data_points):
    plt.annotate(f'{chr(65+i)}', (point[0] + 0.1, point[1] + 0.1), fontsize=12)

plt.title('K-Means Clustering (k=2)')
plt.xlabel('X-coordinate')
plt.ylabel('Y-coordinate')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
from sklearn.cluster import KMeans

# New data points (1D, so reshape for KMeans)
data_points_new = np.array([2, 4, 10, 12, 3, 20, 30, 11, 25]).reshape(-1, 1)

# New initial centroids (1D, so reshape)
initial_centroids_new = np.array([2, 20]).reshape(-1, 1)

k = 2 # Number of clusters

print("New Data Points:")
print(data_points_new.flatten())
print("\nNew Initial Centroids:")
print(initial_centroids_new.flatten())

# Initialize KMeans with the specified number of clusters and initial centroids
kmeans_new = KMeans(n_clusters=k, init=initial_centroids_new, n_init=1, random_state=42)

# Fit the model to the data points
kmeans_new.fit(data_points_new)

# Get the cluster assignments for each data point
labels_new = kmeans_new.labels_

# Get the final centroids
final_centroids_new = kmeans_new.cluster_centers_

print("\nCluster Assignments:")
for i, point_label in enumerate(labels_new):
    print(f"Point {data_points_new[i][0]}: Cluster {point_label}")

print("\nFinal Centroids:")
for i, centroid in enumerate(final_centroids_new):
    print(f"Centroid {i}: {centroid[0]:.2f}")

# Divide the dataset into 2 parts
cluster_0_new = data_points_new[labels_new == 0]
cluster_1_new = data_points_new[labels_new == 1]

print("\nDataset Divided into 2 Parts:")
print(f"Cluster 0: {cluster_0_new.flatten()}")
print(f"Cluster 1: {cluster_1_new.flatten()}")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 4))

# Plot data points for Cluster 0
plt.scatter(cluster_0_new.flatten(), np.zeros_like(cluster_0_new.flatten()),
            s=100, c='red', label='Cluster 1 Data Points', alpha=0.7)
# Plot data points for Cluster 1
plt.scatter(cluster_1_new.flatten(), np.zeros_like(cluster_1_new.flatten()),
            s=100, c='blue', label='Cluster 2 Data Points', alpha=0.7)

# Plot final centroids
# Give centroids a slightly different y-coordinate to make them clearly visible above the data points
plt.scatter(final_centroids_new[0][0], 0.1, s=200, c='darkred', marker='*', label=f'Centroid 1 ({final_centroids_new[0][0]:.2f})')
plt.scatter(final_centroids_new[1][0], 0.1, s=200, c='darkblue', marker='*', label=f'Centroid 2 ({final_centroids_new[1][0]:.2f})')

# Annotate data points
for i, point in enumerate(data_points_new.flatten()):
    plt.annotate(str(point), (point, -0.05), textcoords="offset points", xytext=(0,-10), ha='center', fontsize=9)

plt.title('K-Means Clustering for 1D Data')
plt.xlabel('Data Value')
plt.yticks([]) # Hide y-axis ticks as it's a 1D visualization
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
